```python
import pandas as pd
```

```python
df = pd.read_csv('mobile_price_data.csv')
df
```

```python
df.info()
```

```python
df['mobile_price'] =pd.to_numeric(df['mobile_price'].str.replace(r'[^\d.]', '', regex=True))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for cols in df.columns:
    if df[cols].dtype == object:
        df[cols] = encoder.fit_transform(df[cols])
```

```python
df = df.drop('bluetooth',axis = 1)
```

```python
df.convert_dtypes()
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.corr(),annot = True)
```

```python
sns.distplot(df['mobile_price'])
```

```python
sns.catplot(data = df, x = 'int_memory', y = 'mobile_price')
```

```python
x = df.drop('mobile_price',axis = 1)
y = df['mobile_price']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
y_pred3 = dtr.predict(x_test)
mse3 = mean_squared_error(y_test,y_pred3)
mse3
```
